```python
import requests
from bs4 import BeautifulSoup
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
import json
```

```python
# If you want to open Chrome
driver = webdriver.Chrome()

driver.get("https://rpggeek.com/login")
username = driver.find_element("id","inputUsername")
password = driver.find_element("id","inputPassword")
username.send_keys("radhikesh97")
password.send_keys("Board@97")
driver.find_element(By.CSS_SELECTOR, ".btn.btn-lg.btn-primary").click()
```

```python
# --- pages ---

urls = []

url = 'https://rpggeek.com/browse/rpg/page/{}/'
 
for page in range(1, 145):
    print('\n--- PAGE:', page, '---\n')
 
    # read page with list of posts
    driver.get(url.format(page))
    soup = BeautifulSoup(driver.page_source, 'html.parser') # 'lxml'
    
    all_urls = soup.find_all(class_ = 'primary')
    
    for a_tag in all_urls:
        urls.append(a_tag['href'])
```

```python
urls_d = {}
urls_d['urls'] = urls
urls_json = json.dumps(urls_d)
with open('rpgg_urls.json','w') as f:
    f.write(urls_json)
```

```python
with open('rpgg_urls.json','r') as f:
    urls_rd = json.loads(f.read())
```

```python
urls_new = urls_rd['urls']
```

```python
import time
base_url = 'https://rpggeek.com'
d = {'results':[]}
count = 1
for url in urls_new:
    if(count%100 == 0):
        print('Reached url: ', count)
#     response = requests.get('https://boardgamegeek.com/boardgame/224517/brass-birmingham')
    response = requests.get(base_url + url)
    soup = BeautifulSoup(response.text, 'html.parser')
    title = soup.find("meta", attrs={"name" : "title"})['content']
    description = soup.find("meta", attrs={"name" : "description"})['content']
    rate = float(soup.find("span", attrs={"property" : "v:average"}).string)

    rpgsystem = soup.find(lambda tag: tag.name == "a" and 'href' in tag.attrs and '/rpgsystem/' in tag['href'])
    publisher = ""
    system_link = ""
    if(rpgsystem is not None):
        publisher = rpgsystem.string
        system_link = rpgsystem['href']

#     system_link = soup.find(lambda tag: tag.name == "a" and 'href' in tag.attrs and '/rpgsystem/' in tag['href'])['href']

    rpg_data = {"name": title, "description": description, "rate": rate, "type": "RPG Game", "rpgsystem" : publisher, "system_link" : system_link}
    d['results'].append(rpg_data)
    #         print(info)
    if(count % 100 == 0):
        if(count != 100):
            with open('rpgg.json','r') as f:
                old_d = json.loads(f.read())
                old_list = old_d['results']
                current_list = d['results']
                new_list = old_list + current_list
                d['results'] = new_list
        json_data = json.dumps(d)
        with open('rpgg.json','w') as f:
            f.write(json_data)
        d = {'results':[]}
        if(count == 10000):
            break
    count += 1
```
