```python
import requests
from bs4 import BeautifulSoup
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
import json
```

```python
# If you want to open Chrome
driver = webdriver.Chrome()

driver.get("https://boardgamegeek.com/login")
username = driver.find_element("id","inputUsername")
password = driver.find_element("id","inputPassword")
username.send_keys("radhikesh97")
password.send_keys("Board@97")
driver.find_element(By.CSS_SELECTOR, ".btn.btn-lg.btn-primary").click()
```

```python
# --- pages ---

urls = []

url = 'https://boardgamegeek.com/browse/boardgame/page/{}/'
 
for page in range(1, 501):
    print('\n--- PAGE:', page, '---\n')
 
    # read page with list of posts
    driver.get(url.format(page))
    soup = BeautifulSoup(driver.page_source, 'html.parser') # 'lxml'
    
    all_urls = soup.find_all(class_ = 'primary')
    
    for a_tag in all_urls:
        urls.append(a_tag['href'])
```

```python
urls_d = {}
urls_d['urls'] = urls
urls_json = json.dumps(urls_d)
with open('bgg_urls.json','w') as f:
    f.write(urls_json)
```

```python
with open('bgg_urls.json','r') as f:
    urls_rd = json.loads(f.read())
```

```python
urls = urls_rd['urls']
```

```python
import time
base_url = 'https://boardgamegeek.com'
d = {'results':[]}
count = 1
for url in urls:
    if(count%1000 == 0):
        print('Reached url: ', count)
#     response = requests.get('https://boardgamegeek.com/boardgame/224517/brass-birmingham')
    response = requests.get(base_url + url)
    soup = BeautifulSoup(response.text, 'html.parser')
    data = soup.find_all('script')
    for script in data:
        val = script.string
        if(val is not None and 'GEEK.geekitemPreload' in script.string):
            details = val.split('GEEK.geekitemPreload = ')[1].split(';\n')[0]
    #         print(details)
            info = json.loads(details)
            include_list = ['subtypename','rankinfo','label','polls','stats','name', 'alternatename', 'yearpublished', 'minplayers', 'maxplayers', 'minplaytime', 'maxplaytime', 'minage','short_description','description','wiki','website','links']
            temp_d = {key: info['item'][key] for key in include_list if key in info['item']}
            d['results'].append(temp_d)
    #         print(info)
    if(count % 10000 == 0):
        if(count != 10000):
            with open('bgg.json','r') as f:
                old_d = json.loads(f.read())
                old_list = old_d['results']
                current_list = d['results']
                new_list = old_list + current_list
                d['results'] = new_list
        json_data = json.dumps(d)
        with open('bgg.json','w') as f:
            f.write(json_data)
        d = {'results':[]}
        time.sleep(10)
    count += 1
    #         print(info['item']['stats'])
```
